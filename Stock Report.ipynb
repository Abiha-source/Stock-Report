```python
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf

# Get Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)

# Plot using Matplotlib
plt.figure(figsize=(12, 6))
plt.plot(tesla_data['Date'], tesla_data['Close'], color='red', label='Tesla Closing Price')
plt.title('Tesla Stock Price Over Time', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```

```python
import yfinance as yf
import pandas as pd

# Get GameStop stock data
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)
import matplotlib.pyplot as plt
```

```python
# Define a plotting function similar to Plotly-style
def make_graph(data, stock, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'], data['Close'], label=f'{stock} Closing Price', color='blue')
    plt.title(title, fontsize=16)
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
make_graph(gme_data, 'GameStop', 'GameStop Stock Price Over Time')
```
